### **A Brief Introduction to Python**

Python is a **multi-paradigm** language, i.e., you can write programs that are
**procedural**, **object-oriented**, or **functional** in nature.

* Most programming languages are **procedural**: programs are lists of instructions
that tell the computer what to do with the programs input. E.g., C and Pascal.



* **Object-oriented** programs manipulate collections of objects. Objects have internal state and support methods that query or modify this internal state in some way.
Smalltalk and Java are object-oriented languages. C++ and Python are
languages that support object-oriented programming, but dont force the use of
object-oriented features.

* **Functional** programming decomposes a problem into a set of functions. Ideally,
functions only take inputs and produce outputs, and dont have any internal state that affects the output produced for a given input. Well-known functional
languages include Lisp, Scheme, ML, etc.



```python
# Procedural


mylist = [10, 20, 30, 40]

# modularization
def sum_the_list(mylist):
    res = 0
    for val in mylist:
        res += val
    return res

print(sum_the_list(mylist))
```

```python
# Functional

# Functools contains a collection of higher-order functions
import functools as ft

mylist = [10, 20, 30, 40]

# Recursive Functional approach
def sum_the_list(mylist):

    if len(mylist) == 1:
        return mylist[0]
    else:
        return mylist[0] + sum_the_list(mylist[1:])


# -- lambda is a keyword in Python for defining anonymous functions
# -- reduce(fun, seq) is used to apply a particular
#    function passed in its argument to all of the list elements
#    mentioned in the sequence passed along

print(ft.reduce(lambda x, y: x + y, mylist))
```

```python
# Functions as first-class citizens can be passed as
# arguments and can also be returned as results

# take f and x as arguments and return f(x)
def funcfunc1(f, list):
     return f(list)

funcfunc1(sum_the_list, mylist)



# take f and g as arguments and return f or g
def add(x,y):
  return x+y

def sub(x,y):
  return x-y

def funcfunc2(f, g, code):
     if (code == 0):
         return f
     elif (code == 1):
         return g

print("funcfunc1: ", funcfunc1(sum_the_list, mylist))

print("funcfunc2", funcfunc2(add, sub, 0))
funcfunc2(add, sub, 1)(12, 13)

```

```python
# Object-oriented

class Employee:
   """Class's docstring description
      always goes here"""

   # Class variable
   empCount = 0

   #----------------------------------------------------------
   # __the CONSTRUCTOR has 2 leading and trailing underscroes__
   # Double Leading and Trailing Underscore(__var__): Indicates special methods
   # defined by the Python language.
   # Avoid this naming scheme for your own attributes.

   def __init__(self, name, salary):

      # Call a private method
      # __ triggers name mangling when used in a class context to aid the
      #    interpreter or compiler
      self.__set(name, salary)

      # same as self.empCount
      Employee.empCount += 1

   #----------------------------------------------------------
   # A __PRIVATE method has 2 leading underscores
   def __set(self, name, salary):

     # The instance variables do not need to be declared in advance
     self.name = name
     self.salary = salary

   #----------------------------------------------------------
   # A PUBLIC method
   def displayCount(self):
     print ("Total Employee %d" % Employee.empCount)  # formatted output

   #----------------------------------------------------------
   # Another PUBLIC method
   def displayEmployee(self):
      print ("Name : ", self.name,  ", Salary: ", self.salary)




# object instantiation
e1 = Employee('Sandy', 120000)
e2 = Employee('Mark', 100000)
e3 = Employee('Amy', 87000)


# method invocation
e1.displayEmployee()


# accessing static class variables
print(e2.empCount)


# accessing class docstring
print(e3.__doc__)


# dictionary represention for class
print(Employee.__dict__)


print(e1.__dict__)

```

execution of the program.

# **Python is Block-Oriented**

* A class of high-level languages in which a program is made up of **blocks**  which may include nested blocks as components, such nesting being repeated to any depth.
* A **block** consists of a sequence of statements and/or blocks that is treated as a single unit in its enclosing block.

```python
# The recursive facorial function

def factorial(n):
     if (n < 0):
          return -1
     elif (n==0):
          return 1
     else:
          return n * factorial(n-1)


# test the function
for i in range(1,10,1):
    print(i, factorial(i))

```

### **Python Dictionary**


* A **dictionary** is a collection that allows us to store data in key-value pairs

* Create dictionaries by placing key:value pairs inside curly brackets {}, separated by commas.

```python
# examples of dictionary

# creating a dictionary
country_capitals = {
  "United States": "Washington D.C.",
  "Italy": "Roam",
  "England": "London"
}

# printing the dictionary
print(country_capitals)



# get dictionary's length
print(len(country_capitals))



# Access the value of a dictionary item by placing the key inside square brackets
print(country_capitals["United States"])



# Python dictionaries are mutable
country_capitals["Italy"] = "Rome"
print(country_capitals)



# add an item with "Germany" as key and "Berlin" as its value
country_capitals["Germany"] = "Berlin"
print(country_capitals)



# delete item having "Italy" key
del country_capitals["Italy"]
print(country_capitals)



# Iterating Through a Dictionary
for country in country_capitals:
    capital = country_capitals[country]
    print(capital)



# membership test
# country_capitals["France"]      ERROR
country_capitals.__contains__('France')



# list keys
country_capitals.keys()


# list values
country_capitals.values()



# sort keys
sorted(country_capitals.keys(), reverse=False)


```
